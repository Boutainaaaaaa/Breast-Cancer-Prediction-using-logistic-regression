```python
#importer les données – utilisation de la librairie pandas
import pandas
dataf = pandas.read_excel("breast-w-excelNombre (1).xlsx",sheet_name = 0)
#import libraries
import pandas as pd
import seaborn as sns
```

```python
sns.countplot(dataf['Class'],label="count")
```

```python
# label encoding(convert the value of M and B into 1 and 0)
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
dataf['Class']=labelencoder.fit_transform(dataf['Class'])
```

```python
dataf['Class'].unique()
```

```python
dataf
```

```python
#Splitting the dataset into independent and dependent datasets 
X=dataf[['Clump_Thickness','Cell_Size_Uniformity','Cell_Shape_Uniformity','Marginal_Adhesion','Single_Epi_Cell_Size','Bare_Nuclei','Bland_Chromatin','Normal_Nucleoli','Mitoses']]
X.head()
```

```python
Y=dataf['Class']
Y.head()
```

```python
#Splitting datasets into training(75%) and testing(25%)
from  sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.25)
```

```python
# train logistic regression model 
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train,Y_train) 
```

```python
# make predictions based on X_test and store it to predictions
predictions=logreg.predict(X_test) 
```

```python
#plot confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
cm=confusion_matrix(Y_test,predictions)
print(cm)
sns.heatmap(cm,annot=True)
```

```python
#get accuracy score for model
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,predictions))
```
